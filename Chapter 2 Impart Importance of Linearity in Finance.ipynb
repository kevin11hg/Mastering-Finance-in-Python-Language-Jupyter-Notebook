## CAPM Model

<big>The CAPM model measures the *relationship between the **returns*** of **risk** and stock returns for every stock in the portfolio basket. 

By outlining the $\Sigma$ sum of this relationship,

**we obtain weights** or /combinations of risky securities that ***produces*** *the **portfolio** risk* for the **lowest @every level** of *portfolio return.* 

An investor **who wishes** to **receive a particular return** would own one such ***combination*** of an ***optimal*** ***portfolio*** that
***provides*** the least risk ***possible***. The combinations of optimal portfolios lie along a line called the efficient frontier.

#  Efficient Frontier & Market Portfolio

<blockquote> Curve of combinations of optimal portfolios</blockquote>

# Arbitrage Pricing Theory Model

<big>
    
CAPM suffers from several limitations:
    
#### 1. Use of mean-variance $\mu-$ framework

#### 2. Fact / factor that *returns* are captured by one *risk* factor -- market risk
    
<br/>  
    
Well-diversified portfolio, the unsystematic risk of various stocks cancels out -- essentially eliminated
</big>

<big>
<h1>Arbitrage Pricing Theory (APT)</h1> model was put forward to address these shrotcomings and offers a general approach of determining the asset prices other than the mean and variances
Well-diversified portfolio, the unsystematic risk of various stocks cancels out -- essentially eliminated
<h3>Assumptions</h3>
<br/>
APT model assumes that security returns are generated according to multiple factor models, which consist of a linear combmination of several systematic risk factors.
    
Such factors could be 
+ inflation rate
+ GDP growth rate
+ real interest rates, or
+ dividends

<br/>  
    
Equilibrium Equation and Pricing Assets According to APT model is as follows:
# $E[R_i] = \alpha_i + \beta_{i,1}F_1 + \beta_{i,2}F_2 +\cdot\cdot\cdot$
# $+ \beta_{i,j}F_j$
    
Where, Here 
+ $ E[R_i]$ is the expected *rate of return* on security $i$
+ $\alpha_i$ is the expected return on stock $i$ if all factors are negligble
+ $\beta_{i,j}$ is the sensitivity of the $i$th asset to the $j$th factor
+ $F_j$ is the value of the $j$th factor influencing the return on stock $i$
</big>

## A Simple Linear Optimization problem

<big>Suppose we are interested in investing in two securities,

$x$ and $y$

We would like to find out the

Actual # number of units to invest for every 3 units of security $X$ and 2 units of security $Y$,

S.t.

total # number of units invested is maximized as
Possible

However there are 


Certain constrainst on our

Investment Strategy:

+ For every 2 units of security $X$ invested and 1 unit of security $Y$ invested, the total volume must not exceed 100
+ For every unit of security $X$ and $Y$ invested, total voume must not exceed 80
+ Total volume allowed to invest in seurity $X$ must not exceed 40
+ Short selling forbidden for both securities


<b>Maximization</b> problem can be

<b>Mathematically</b> represented as follows

Maximize: $f(x,y) = 3x + 2y$

$Subject to:$

$2x + y \leq 100$

$x + y \leq 80$

$x \leq 40$

$x \geq 0, y\geq 0$
</big>

<big>By plotting constraints on an $x$ by $y$ graph,
    
Set of feasible solutions is shown in the shaded area:
    
</big>

![](CH-2-MATH-MAX.png)

## TRANSLATED

```python
"""A simple linear optimization problem with 2 variables"""

import pulp
x = pulp.LpVariable("x", lowBound=0)
y = pulp.LpVariable("y", lowBound=0)

problem = pulp.LpProblem("A simple maximization objective", 
                         pulp.LpMaximize)

problem+= 3*x + 2*y, "The objective function"
problem+= 2*x + y <= 100, "1st constraint"
problem+= x + y <= 80,  "2nd constraint"
problem+= x <= 40,   "3rd constraint"


problem.solve()
```

<big>
    
<pre>LpVariable</pre> function defines a variable to be solved
    
<pre>LpProblem</pre> function initializes the problem with a text description of the problem and the type of optimization, which in this case is the maximization method
    
Finally, the <pre>solve</pre> function is called to begin performing linear optimization
    
Each variable value is printed to show values solved for us by optimizer

Following output is generated when the code runs:
</big>

```python
print("Maximization Results: ")

for variable in problem.variables():
    
    print(variable.name + "=" + str(variable.varValue))
```

<big>
       
The results show that obtaining the maximum value of 180 is POSSIBLE
    
when value of $x$ is 20 and $y$ is 60 while
    
fulfilling given set of constraints
</big>

## Example of Integer Programming Model w/ Binary Conditions

<big>
Suppose we must go for 150 contracts in a particular OTC over-the-counter exotic security from three dealers. Dealer $X$ quoted \$500 per contract plus handling fees of \$4,000, regardless of the number of contracts sold, regardless of the number of contracts. Dealer $Y$ charges \$450 per contract plus a transaction fee of \$2,000. Dear $Z$ charges \$450 per contract plus a fee of \$6,000
    
Dealer $X$ will sell at most 100 contracts, dealer $Y$ at most 90, and dealer $Z$ at most 70
    
Minimum transaction volume from any dealer is 30 contracts if any are transacted with that dealer
    

</big>

### How should we minimize the cost of purchasing 150 contracts?

<big>Using the <pre>pulp</pre> package, let's set up the required variables</big>

```python
""" An example of implementing an integer progarmming model w/ binary conditions"""
import pulp


dealers = ["X","Y","Z"]

variable_costs = {"X": 500,
                  "Y": 350,
                  "Z": 450}

fixed_costs =  {"X": 4000,
                "Y": 2000,
                "Z": 6000}

# Define PuLP variables to solve

quantities = pulp.LpVariable.dicts("quantity",
                                  dealers,
                                  lowBound=0,
                                  cat=pulp.LpInteger)


is_orders = pulp.LpVariable.dicts("orders",
                                 dealers,
                                 cat=pulp.LpBinary)
```

<big><pre>dealers</pre> variable contains the dictionary identifiers that are used to reference lists and dictionaries later on
    
The <pre>variable_costs</pre> and  <pre>fixed_costs</pre> variables are dictionaries that contain their respective contract cost and fees charged by each dealer
    
PuLP <b>solver solves</b> for the <b>values</b> of <pre>quantities</pre> and  <pre>is_orders</pre>, which are defined by the <pre>LpVariable</pre> function
    
<pre>dicts</pre> function tells PuLP to treat the assigned variable as a dictionary object, using the <pre>dealers</pre> variable for referencing

Note that the <pre>quantities</pre> variable has a lower boundary of 0 that prevents us from entering a short position in any securities
    
<pre>is_orders</pre> values are treated as binary objects,
    
Indicating whether we should enter into a transaction with any of the dealers
</big>

# 01/07/2020

<big>
<h3>What is the best approach to modeling this integer programming problem?</h3>
    
At first glance, this seems fairly straightforward by applying this equation:
    
#### $ Minimize \text{ } \sum\limits_{i=x}^{i=z} IsOrder_i \text{ } [variablecost_i  \text{ } \times quantity_i + fixed cost_i]$
    
Where

$\begin{align*}
  IsOrder_i   &= \begin{cases} 1, & \text{if buying from dealer i} \\ 0, & \text{otherwise} \end{cases} \\ \\30 &\leq quantity_x \leq 100 \\ 30 &\leq quantity_y \leq 90 \\30 &\leq quantity_z \leq 70 \\\sum\limits_{i=x}^{i=z}&quantity_i = 150 
\end{align*}$
    

</big>

<big>

Equation <b>simply states</b> that we want to minimize the total costs with the
    
binary variable $IsOrder_i$, determining whether to 
    
Account with buying from a sepcific dealer should we choose to
    
Let's implement this model in Python
</big>


```python
"""
Example of implementing an integer programming model with 
binary
variables the wrong way
"""
# Initialize the model with constraints
model = pulp.LpProblem("A cost minimization problem",
                    pulp.LpMinimize)
model += sum([(variable_costs[i] * quantities[i] + fixed_costs[i])*is_orders[i] for i in dealers]),\
"Minimize portfolio cost"

model += sum([quantities[i] for i in dealers]) == 150,\
"Total contracts required"

model += 30 <= quantities["X"] <= 100, "Boundary of total volume of X "
model += 30 <= quantities["Y"] <= 90, "Boundary of total volume of Y"
model += 30 <= quantities["Z"] <= 70, "Boundary of total volume of Z"

model.solve()
```

<big>As it turned out, we were trying to perform multiplication on two unknown variables, <pre>quantities</pre>  and <pre>is_order</pre>, which unknowingly led us to perform nonlinear programming. Such are the pitfalls encountered when performing integer programming
</big>

## A different / new approach with Binary Conditions
<big>Another method of formulating the minimization objective is to place all unknown variables in a linear fashion such that they are additive:</big>

### $Minimize \sum\limits_{i=x}^{i=z} variable cost_i \times quantity_i + fixed cost_i \times IsOrder_i$

<big>Comparing with previous objective equation,
we would <b>obtain</b> the same fixed
    
costs Values.
    
    
    
    
However, the unknown variable $quantity_i$ is required to be solved as a function of $IsOrder_i$ such that
    
constraints are stated as follows:
    
$\begin{align*} IsOrder_i \times 30  &\geq quantity_x \leq IsOrder_i \times 100 \\ IsOrder_i \times 30  &\leq quantity_y \leq IsOrder_i \times 90 \\ IsOrder_i \times 30  &\leq quantity_z \leq IsOrder_i \times 70 \\
\end{align*}$

Let's
    
Apply these Formulas in Python

</big>

```python
"""
This is an example of implementing an IP model
with binary variables the correct way (in Christ!)
"""
### Initialize model /w+ constraints

model = pulp.LpProblem("A cost minimization problem",
                      pulp.LpMinimize)
model += sum([variable_costs[i] * quantities[i] + fixed_costs[i] * is_orders[i] for i in dealers]),\
"Minimize portfolio cost"
model += sum([quantities[i] for i in dealers]) == 150,\
"Total contracts required"


model += is_orders["X"]*30 <= quantities["X"] <=\
is_orders["X"]*100, "Boundary of total volume of X"
model += is_orders["X"]*30 <= quantities["Y"] <=\
is_orders["Y"]*90, "Boundary of total volume of Y"
model += is_orders["Z"]*30 <= quantities["Z"] <=\
is_orders["Z"]*70, "Boundary of total volume of Z"
model.solve()
```

<h3>
What happens when we try to run the solver?
</h3>



```python
print("Minimization Results:")

for variable in model.variables():
    print("variable", "=", variable.varValue)


print("Model Objective %sTotal cost: " % model.objective )
print("Total cost: %s" % pulp.value(model.objective))
```

## LU decomposition

<b>LU decomposition</b> also known as AKA

<b>Lower Upper factorization</b> is one of the methods of 

solving square systems of linear equations

Name implies? LU decomposition decomposes matrix $A$ into a product of two matrices:

lower triangular matrix $L$ and an upper triangular matrix $U$



Decomposition can be represented as follows:

$A = LU$

$ \begin{bmatrix}
a & b & c\\
d & e & f\\
g & h & i
\end{bmatrix} = \begin{bmatrix}
l_{11} & 0 & 0\\
l_{21} & l_{22} & 0\\
l_{31} & l_{32} & l_{33}
\end{bmatrix} \times \begin{bmatrix}
u_{11} & u_{12} & u_{13}\\
0 & u_{22} & u_{23}\\
0 & 0 & u_{33}
\end{bmatrix},  $ 

<big>Here

We can see
    
$a = l_{11}u{11}, b = l_{11}u{12}$

Lower triangular matrix is a matrix that contains values in its lower triangle with the remaining uper triangle populated with zeroes

Converse is true for an upper triangular matrix

Definite advantage of the LU decomposition method over the Cholesky decompositino methods is that it works for any square matrices

Latter method only works for symmetric and positive definite matrices

Remember pervious example of a 3 by 3 matrix A

This time, though, we will use the <pre>linalg</pre> package of the <pre>SciPy</pre> module for the LU decomposition

</big>

```python
""" LU decomposition with SciPy """
import scipy.linalg as linalg
import numpy as np

A = np.array([[2,1,1],
             [1,3,2],
             [1,0,0]])
B = np.array([4,5,6])
LU= linalg.lu_factor(A)

x = linalg.lu_solve(LU, B)
```

## QR Decomposition

<big>QR decomposition, also known as the QR factorization, is another method of

solving linear systems of equations using matrices, very much like the LU decomposition

Equation to solve $Ax = B$, where matrix

$A = QR$ 

Except in this case $A$ is a product of an orthogonal matrix $Q$ and upper triangular matrix $R$

QR algorithm commonly used to solve the linear least squares problem

Orthogonal matrix exhibits following properties:

+ Square Matrix
+ Multiplying orthogonal matrix by its transpose returns identity matrix:

$QQ^T = Q^TQ = 1$

+ Inverse of orthogonal matrix equals its transpose:

$Q^T = Q^-1$
    
Identity matrix is also a square matrix with its main diagonal containing ones and zeroes elsewhere
</big>

<big>We can
    
Now restate the problem $Ax = B$ as follows:
       
$QRx = B$

$Rx = Q^-1B \text{ or } Rx = Q^TB$
    
Use same variables in LU decomposition example example,
    
We will use the <pre>qr</pre> function of <pre>scipy.linalg</pre> to compute our values of $Q$ and $R$, and let the
    
variable $y$ 
    
Represent our value of $BQ^T$ with following code:

```python
""" QR decomposition with Scipy """
import scipy
import scipy.linalg as linalg
import numpy as np

A = np.array([
    [2,1,1],
    [1,3,2],
    [1,0,0]
])

B = np.array([4,5,6])

Q, R = scipy.linalg.qr(A) # QR decomposition
y = np.dot(Q.T,B) # Let y = Q'.B
x = scipy.linalg.solve(R, y) # Solve Rx = y
```

### Jacobi method

<big>Jacobi method
    
Solves a system of linear equations iteratively along its diagonal elements
    
Iteration procedure terminates when the solutino converges
    
Again the equatino to solve is in the form of $Ax = B$, where $A$ can be decomposed into a diagonal component $D$ and remainder $R$ such that $A = D + R$ 
    
Let's
    
Look at the example of a 4 by 4 matrix $A$
    
$A = \begin{bmatrix}
a & b & c & d \\
e & f & g & h\\
i & j & k & l\\
m & n & o & p
\end{bmatrix} = \begin{bmatrix}
a & 0 & 0 & 0 \\
0 & f & 0 & 0\\
0 & f & 0 & 0\\
0 & 0 & 0 & p
\end{bmatrix} +  \begin{bmatrix}
0 & b & c & d \\
e & 0 & g & h\\
i & j & 0 & l\\
m & n & o & 0
\end{bmatrix} $ 

<big>Solution is then obtained iteratively as follows:
    
$ Ax = B $
    
    
$ (D + R)x = B $
    
    
$ Dx = B - Rx $
    
    
$ x_{n + 1} = D'(B - Rx_n) $
    
    
As opposed to the Gauss-Siedel method, the value of $x_n$ in the Jacobi method is needed during each iteration in order to compute $x_{n + 1}$ and cannot be overwritten
    
This is would take up twice amount of storage
    
However, the computations for each element can be done in parallel, which is useful can be done in parallel, which is useful for faster computations
    
    
If matrix $A$ is <b>strictly irreducibly</b>
    
<b>diagonally dominant</b>, this method is
    
<b>guaranteed</b> to <b>converge</b>
    
A
    
+ strictly irreducibly
+ diagonally dominant
    
matrix is one where the absolute diagonal element in every row is greater than the sum of the absolute values of other terms
    


<big>In some circumstances, the
    
Jacobi method
    
Can <b>converge even</b> if these <b>conditions</b> are <i>not</i> met 
    
Python code is
    
Given as follows:

```python
""" Solve Ax = B with Jacobi method """

import numpy as np

def jacobi(A, B, n, tol=1e-10):
    
    # Initializes x with zeroes with same shape and type as B
    x = np.zeros_like(B)
    
    for it_count in range(n):
        x_new = np.zeros_like(n)
        for i in range(A.shape[0]):
            
            
            s1 = np.dot(A[i, :i], x[:i])
            s2 = np.dot(A[i, i + 1:], x[i + 1:])
            x_new[i] = (B[i] - s1 - s2)  /A[i,i]
            
        if np.allclose(x, x_new, tol):
            break

        x = x_new

    return x
```

<big>Consider same matrix values in Cholesky decomposition exampke

We will use 25 iterations in our <pre>jacobi</pre> function to find the values of $x$

```python
A = np.array([[10,-1, 2,0],
             [-1, 11, -1, 3],
             [2, -1, 10, -1],
             [0, 3, -1, 8]])
B = np.array([6, 25,-11,15])

n = 25
```

<big>After initializing the values

We <b>can call now</b> the function and solve for $x$:

```python
x = jacobi(A, B, n)
print("x = ", x)
```

<big>We <b>solved</b> for the values of $x$, which are <b>similar</b> to the answers from the Cholesky decomposition

## Gauss-Seidel method

<big>Gauss works very much like the Jacobi method another a-new way
    
<b>solving a square system</b> of lineaer equations using an iterative procedure with the equation in the form of $Ax = B$
    
Here the matrix $A$ is decomposed as
    
$A = L + U$
where the matrix $A$ is a sum of a lower triangular matrix $L$ and an upper triangular matrix $U$
    
    
Let's take a
    
Look at the example of a 4 by 4 matrix $A$:
    
$A = \begin{bmatrix}
a & b & c & d \\
e & f & g & h\\
i & j & k & l\\
m & n & o & p
\end{bmatrix} = \begin{bmatrix}
a & 0 & 0 & 0 \\
e & f & 0 & 0\\
i & j & k & 0\\
m & n & o & p
\end{bmatrix} +  \begin{bmatrix}
0 & b & c & d \\
0 & 0 & g & h\\
0 & 0 & 0 & l\\
0 & 0 & 0 & 0
\end{bmatrix} $ 

<big> Solution is then obtained iteratively as follows:
    
$Ax = B$
    
$(L + U)x = B$
    
$Lx = B - Ux$
    
$x_{n + 1} = L^{-1}(B-Ux_n)$
    
Using a lower triangular matri $L$, where zeroes fill u pupper triangle, the
    
elements of $x_n$ can be overwritten in each iteration in oder to compute $x_{n+1}$
    
Results in advantage of needing half the storage required when using the acjobi method
    
Rate of convergence in the Gauss-Seidel method largely lies in the properties of matrix $A$, especially if matrix $A$ is needed to be strictly diagonally dominant or symmetric positive definite
    
Even if these conditions are not met, the Gauss-Seidel
    
method may converge
    
Python implementation of the Gauss-Seidel
    
method is given as follows:
    


```python
""" Solve Ax = B Gauss-Seidel method """
import numpy as np

def gauss(A, B, n, tol=1e-10):
    L = np.tril(A) # Returns lower triangular matrix of A
    U = A - L # Decompose A = L + U
    L_inv = np.linalg.inv(L)
    x = np.zeros_like(B)
    
    for i in range(n):
        Ux = np.dot(U, x)
        x_new = np.dot(L_inv, B - Ux)
        
        if np.allclose(x, x_new, tol):
            break
            
    x = x_new
    
    return x
```

<big>Here

<pre>tril</pre> function of NumPy returns lower triangular matrix $A$ from which

<b>We can</b> find the lower triangular matrix $U$
    
Plugging the remaining values
into $x$ iteratively would lead us to the solution below with some tolerance defined by 
<pre>tol</pre>
    
Let's consider the same matrix values in the Jacobi method and Cholesky decomposition example

    
    
We will use a maximum of 100 iterations in our <pre>gauss</pre> function to find the values of $x$ as follows:

```python
A = np.array([[10,-1,2,0],
              [-1,11,-1,3],
              [2,-1,10,-1],
              [0,3,-1,8]])
B = np.array([6,25,-11,15])
n = 100
x = gauss(A,B,n)
```

<big>Let's see if our $x$ values match with those as before

```python
print("x =",x)
```

<big>We solved for the values of $x$, which are similar to the answers from the Jacobi method and Cholesky decomposition
